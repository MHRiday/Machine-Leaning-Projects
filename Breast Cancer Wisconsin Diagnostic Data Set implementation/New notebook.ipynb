```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
train,test = train_test_split(data,test_size = .2,random_state = 42)
```

```python
test.shape
```

```python
train.shape
```

```python
train_sets_column = ['concave points_worst','perimeter_worst','radius_worst','area_worst']
```

```python
Train_test_sets_column = ['diagnosis']
```

```python
train_data = train[train_sets_column]
```

```python
train_test_data = train[Train_test_sets_column]
```

```python
test_data = test[train_sets_column]
test_outcomes = test[Train_test_sets_column]
```

```python
train_data.shape
```

```python
train_test_data.shape
```

```python
test_data.shape
```

```python
test_outcomes.shape
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(train_data,train_test_data)
```

```python
y_predict = clf.predict(test_data)
```

```python
accuracy_score(test_outcomes,y_predict)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
clf = RandomForestClassifier(n_estimators=13)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, train_test_data, cv=k_fold, n_jobs=1, scoring=scoring)
print(np.mean(score))
```

```python
from sklearn import svm
clf = svm.LinearSVC()
clf.fit(train_data,train_test_data)
y_predict = clf.predict(test_data)
```

```python
accuracy_score(test_outcomes,y_predict)
```

```python
from sklearn import svm
clf = svm.LinearSVC()
scoring = 'accuracy'
score = cross_val_score(clf, train_data, train_test_data, cv=k_fold, n_jobs=1, scoring=scoring)
print(np.mean(score))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
clf.fit(train_data,train_test_data)
y_predict = clf.predict(test_data)
accuracy_score(test_outcomes,y_predict)
```

```python
accuracy_score(test_outcomes,y_predict)
```

```python
clf = DecisionTreeClassifier(random_state=0)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, train_test_data, cv=k_fold, n_jobs=1, scoring=scoring)
print(np.mean(score))
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(train_data,train_test_data)
y_predict = neigh.predict(test_data)
accuracy_score(test_outcomes,y_predict)
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
scoring = 'accuracy'
score = cross_val_score(neigh, train_data, train_test_data, cv=k_fold, n_jobs=1, scoring=scoring)
print(np.mean(score))
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial').fit(train_data,train_test_data)
clf.fit(train_data,train_test_data)
y_predict = clf.predict(test_data)
accuracy_score(test_outcomes,y_predict)
```

```python
clf = clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial')
scoring = 'accuracy'
score = cross_val_score(clf, train_data, train_test_data, cv=k_fold, n_jobs=1, scoring=scoring)
print(np.mean(score))
print(score)
```
